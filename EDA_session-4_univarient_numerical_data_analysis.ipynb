**Percentale and quantile**
- in the numpy package we have the np.percentile() and np.quartile()

- percentile: if you want to 25p  25

- quantile: q1=25p (0.25)  q2=50p   q3=75p

- assume that a student got 120 marks 95p

- 95% of students has marks below 120



```python
p_wage=visa_df['prevailing_wage']
########## PANDAS  ##############
wage_count = round(p_wage.count(),2)
wage_min = round(p_wage.min(),2)
wage_max = round(p_wage.max(),2)
wage_mean = round(p_wage.mean(),2)
wage_median = round(p_wage.median(),2)
wage_std=round(p_wage.std(),2)

######  NUMPY ####################
wage_25p=round(np.percentile(p_wage,25),2)
wage_50p=round(np.percentile(p_wage,50),2)
wage_75p=round(np.percentile(p_wage,75),2)



list1=[wage_count ,
wage_min,
wage_max ,
wage_mean ,
wage_median,
wage_std,
wage_25p,
wage_50p,
wage_75p ]

index_list=['Count','Min','Max','Mean','Median','std','25p','50p','75p']
table=pd.DataFrame(list1,columns=['prevailing_wage'],
            index=index_list)
table
```

```python
num_cols=visa_df.select_dtypes(exclude='object').columns
dict1={}
for i in num_cols:
    
    count = round(visa_df[i].count(),2)
    minn = round(visa_df[i].min(),2)
    maxx = round(visa_df[i].max(),2)
    mean = round(visa_df[i].mean(),2)
    median = round(visa_df[i].median(),2)
    std=round(visa_df[i].std(),2)

    _25p=round(np.percentile(visa_df[i],25),2)
    _50p=round(np.percentile(visa_df[i],50),2)
    _75p=round(np.percentile(visa_df[i],75),2)

    list1=[count ,minn,maxx ,mean ,medianstd,_25p,_50p,_75p ]
    dict1[i]=list1

index_list=['Count','Min','Max','Mean','Median','std','_25p','_50p','_75p']
numeric_df = pd.DataFrame(dict1,index=index_list)
numeric_df.to_csv('number_df.csv')
numeric_df
```
